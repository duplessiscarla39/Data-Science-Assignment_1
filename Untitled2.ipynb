```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# 1. Setup Parameters
years = np.arange(2026, 2032)
neighborhoods = ["Sea Point", "Green Point", "Gardens", "CBD", "Woodstock",
                 "Observatory", "Camps Bay", "Claremont", "Muizenberg", "Hout Bay"]

# Create a list to store data
data = []

# 2. Generate Base Data
for nb in neighborhoods:
    # Starting values for each neighborhood
    start_str_density = np.random.uniform(2, 5)
    growth_rate = np.random.uniform(1.5, 3.0)
    base_rent = np.random.uniform(12000, 22000)

    for i, year in enumerate(years):
        # STR density increases over time
        str_density = start_str_density + (growth_rate * i)

        # Natural inflation (5% per year)
        natural_growth = (1.05)**i

        # Non-linear displacement factor (Impact of STR on LTR)
        displacement_factor = 1 + (str_density / 15)**2

        # Projected Rent
        projected_rent = base_rent * natural_growth * displacement_factor

        data.append({
            'Year': year,
            'Neighborhood': nb,
            'STR_Density_Pct': str_density,
            'Base_Rent_Baseline': base_rent,
            'Projected_Monthly_Rent_ZAR': projected_rent
        })

df = pd.DataFrame(data)

# 3. Define Regulatory Risk Threshold
# Intervention is likely if STR Density > 10% or Rent Growth > 40% from baseline
def calculate_risk(row):
    # Calculate rent increase percentage for that specific neighborhood relative to 2026
    baseline_rent = df[(df['Neighborhood'] == row['Neighborhood']) & (df['Year'] == 2026)]['Projected_Monthly_Rent_ZAR'].values[0]
    rent_increase_pct = (row['Projected_Monthly_Rent_ZAR'] - baseline_rent) / baseline_rent

    # Risk score: 60% weight on density, 40% weight on rent increase
    risk_score = (row['STR_Density_Pct'] * 0.6) + (rent_increase_pct * 100 * 0.4)
    return risk_score, rent_increase_pct * 100

# Apply calculations
df[['Reg_Risk_Score', 'Rent_Increase_Pct']] = df.apply(
    lambda row: pd.Series(calculate_risk(row)), axis=1
)

# Identify risk category
def categorize_risk(score):
    if score > 15: return "Critical (Immediate Intervention Likely)"
    if score > 10: return "High (Policy Monitoring)"
    return "Moderate"

df['Risk_Category'] = df['Reg_Risk_Score'].apply(categorize_risk)

# Save to CSV
df.to_csv('cape_horizon_synthetic_displacement_2026.csv', index=False)

# 4. Visualization 1: Rent Projection
plt.figure(figsize=(12, 7))
sns.lineplot(data=df, x='Year', y='Projected_Monthly_Rent_ZAR', hue='Neighborhood', marker='o')
plt.axhline(y=25000, color='red', linestyle='--', label='Affordability Crisis Threshold')
plt.title('Cape Horizon Realty: Projected LTR Rent Displacement (2026-2031)', fontsize=14)
plt.ylabel('Average Monthly Rent (ZAR)', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('rent_projection_2026.png')

# 5. Visualization 2: STR Density vs Risk Score
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df[df['Year'] == 2031], x='STR_Density_Pct', y='Reg_Risk_Score',
                hue='Neighborhood', s=200)
plt.axhline(y=15, color='darkred', linestyle='--', label='Regulatory Crackdown Trigger')
plt.title('Regulatory Risk vs. STR Penetration (Forecasted 2031)', fontsize=14)
plt.xlabel('Short-Term Rental Density (% of Housing Stock)', fontsize=12)
plt.ylabel('Regulatory Risk Score', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('regulatory_risk_2031.png')

print(df.head())
```
